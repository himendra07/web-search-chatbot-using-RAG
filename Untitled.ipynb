```python
# Preferred in Jupyter / IPython: use top-level await
import asyncio
from web_scraper import get_urls, fetch_web_pages, encode_url_to_filename
import os

print("CWD:", os.getcwd())
print("Project files:", sorted([f for f in os.listdir('.') if f.endswith('.py') or f.endswith('.txt') or f.endswith('.ipynb')]))

query = "LangChain tutorial"
urls = get_urls(query, num_results=2, provider="duckduckgo")
print("\nDuckDuckGo returned URLs:")
for u in urls:
    print(" -", u)

# run the async fetch using await
async def run_fetch():
    await fetch_web_pages([query], num_results=2, provider="duckduckgo")

await run_fetch()   # <--- top-level await works in Jupyter

print("\nDownloaded files (if any):")
download_dir = "./downloaded"
if os.path.exists(download_dir):
    for fn in os.listdir(download_dir):
        print(" -", fn)
else:
    print(" (no downloaded folder found)")
```
